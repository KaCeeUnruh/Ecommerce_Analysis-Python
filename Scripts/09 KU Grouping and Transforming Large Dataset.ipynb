Aggregate mean of order_number grouped by department_id

```python
# Average number of orders per user across each Instacart department
df_ords_prods_merge.groupby('department_id').agg({'order_number': ['mean']})
```

Create a loyalty flag for existing customers using the transform() and loc() functions

```python
# New column- Max order per customer
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check output with max order column
df_ords_prods_merge.head(20)
```

```python
# Create a loyalty flag
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Count loyalty_flag values
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

Use the loyalty flag to check the basic statistics of product prices for each loyalty category 

```python
# Group data by user_id, find mean product price for each user. mean_price for results
df_ords_prods_merge['mean_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check basic statistics for each customer loyalty group
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Create a spending flag for each user based on average price across all their orders 

```python
# Create spending flags to user ID based on mean product price
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] < 10, 'spending_flag'] = 'Low spender'
df_ords_prods_merge.loc[df_ords_prods_merge['mean_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Count spending_flag values
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

Determine frequent versus non-frequent customers

```python
# Group data by user_id, find median product of days since last order. median_days for results
df_ords_prods_merge['median_days'] = df_ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
# Check output
df_ords_prods_merge.head(20)
```

```python
# Create frequency flags per user_id based on median days since prior order
df_ords_prods_merge.loc[df_ords_prods_merge['median_days'] > 20, 'frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days'] > 10) & (df_ords_prods_merge['median_days'] <= 20), 'frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_days'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Count frequency flags
df_ords_prods_merge['frequency_flag'].value_counts(dropna=False)
```
