# Grouping Data & Aggregating Variables

Import Libraries

```python
# Import libraries 
import pandas as pd
import numpy as np
import os
```

Import Datasets

```python
# Create path shortcut
path = r'C:\Users\kacee\OneDrive\Desktop\Instacart Basket Analysis'
```

```python
# Import pkl orders_products_merged_updatedcolumns
df_ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_updatedcolumns.pkl'))
```

```python
# Check output
df_ords_prods_merge
```

Cleanup Dataset

```python
# Remove _merge column
df_ords_prods_merge = df_ords_prods_merge.drop(columns = ['_merge'])
```

Creating subset to explore data

```python
# Create subset
df = df_ords_prods_merge[:1000000]
```

```python
# Check subset shape
df.shape
```

```python
# Check subset output
df.head()
```

```python
# Group subset by product name
df.groupby('product_name')
```

```python
# Average number of orders per user across each Instacart department using agg()
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Average number of orders per user across each Instacart department
df.groupby('department_id')['order_number'].mean()
```

```python
# Mean, Min, Max orders across each Instacart department
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

Create max_order column and create a loyalty_flag in complete dataset

```python
# New column- Max order per customer
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check output with max order column
df_ords_prods_merge.head(15)
```

```python
# Clear max number of rows that can be displayed
pd.options.display.max_rows = None
```

```python
# Create a loyalty flag
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Count loyalty_flag values
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
